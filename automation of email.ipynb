```python

import sys

import os

import datetime

import smtplib

 #import the following module

from email.mime.multipart import MIMEMultipart

from email.mime.text import MIMEText
from google.colab import drive
drive.mount('/content/drive')
with open('/content/drive/My Drive/schedule.txt', 'r') as f: 
    schedule = f.readlines()
    file_content = [a.strip().split(' ') for a in schedule] #convert file in a list of list
    for i in file_content:
        i.remove('-')
    date_auto =[' '.join(i[0:3])for i in file_content] #getting list of schedule class
    time_auto =[''.join(i[3:5])for i  in file_content] #getting list of schedule time
    class_module = [' '.join(i[5:]) for i in file_content] # geting the list of schedule class module
    value_list = []
    for i in range(len(time_auto)):
        value_1 = []
        value_1.append(time_auto[i])
        value_1.append(class_module[i])
        value_list.append(value_1)
    dict_file_auto = dict(zip(date_auto,value_list))
    schedule_date = list(dict_file_auto.keys())
    format = "%Y %d %B"
    today=datetime.date.today()  # checking the weekday of today 
    if today.strftime("%a") in ['Mo','Tu','We','Th']:
        a=f'This is a reminder mail for your class on {dict_file_auto[(today+datetime.timedelta(days=1)).strftime(format)][1]} at {(today+datetime.timedelta(days=1)).strftime(format)} {dict_file_auto[(today+datetime.timedelta(days=1)).strftime(format)][0]}. Please join the class on time.'
    elif today.strftime("%a") == 'Fr':
        a=f'This is a reminder mail for your class on {dict_file_auto[(today+datetime.timedelta(days=3)).strftime(format)][1]} at {(today+datetime.timedelta(days=3)).strftime(format)} {dict_file_auto[(today+datetime.timedelta(days=3)).strftime(format)][0]}. Please join the class on time.'

        mail_content = a
        #The mail addresses and password
        sender_address = sys.argv[1] # provided as a sys arguments
        sender_pass = sys.argv[2]  # provided as a sys arguments
        receiver_address = ['sbm.iet@gmail.com','sbm.piemr@gmail.com']
        #Setup the MIME
        message = MIMEMultipart()
        message['From'] = sender_address 
        message['To'] = ','.join(receiver_address)
        message['Subject'] = 'OneLearn course series reminder'   #The subject line
        #The body and the attachments for the mail
        message.attach(MIMEText(mail_content, 'plain'))
        #Create SMTP session for sending the mail
        session = smtplib.SMTP('smtp.gmail.com', 587) #use gmail with port
        session.starttls() #enable security
        session.login(sender_address, sender_pass) #login with mail_id and password
        text = message.as_string()
        session.sendmail(sender_address, receiver_address, text)
        session.quit()
        print('Mail Sent')

    


```
